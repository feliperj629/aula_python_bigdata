Função para calcular a média manualmente

```python
# Função para calcular a média manualmente
def media_manual(lista):
    soma = sum(lista)  # Soma de todos os valores na lista
    quantidade = len(lista)  # Número de elementos na lista
    media = soma / quantidade  # Média aritmética
    return media

print(media_manual(temperaturas))
```

Exibindo a mediana de temperaturas:

```python
print("Mediana:", temperaturas.median())
```

Resumo das diferenças:

  **Média**: Soma dos valores dividida pelo número de elementos, sensível a outliers.

  **Mediana**: Valor central após ordenação dos dados, não sensível a outliers.

**Exibindo o desvio padrão:**

O desvio padrão é uma medida que indica o quanto os valores de um conjunto de dados estão dispersos em relação à média. Em outras palavras, ele mostra o quanto, em média, os valores de um conjunto se afastam da média.

```python
print("Desvio padrão:", temperaturas.std())
```

**Gráfico de dispersão:**

O comando df2.sample(frac=.1).plot.scatter(x='longitude', y='latitude') seleciona uma amostra aleatória de 10% dos dados de df2 e cria um gráfico de dispersão com longitude no eixo X e latitude no eixo Y, permitindo visualizar a distribuição espacial dos dados.

```python
df2.sample(frac=.1).plot.scatter(x='longitude', y='latitude')
```

**Gráfico de dispersão com população X valor médio das casas:**

O comando df2.sample(frac=.1).plot.scatter(x='population', y='median_house_value') seleciona uma amostra aleatória de 10% dos dados de df2 e cria um gráfico de dispersão com população no eixo X e valor médio das casas no eixo Y, permitindo visualizar a relação entre o tamanho da população e o valor médio das casas nos dados.

```python
df2.sample(frac=.1).plot.scatter(x='population', y='median_house_value')
```
