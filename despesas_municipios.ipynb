```python
#Calculando a soma, a média e a Mediana de uma coluna particular
soma_data = data["VALUE"].sum()

media_data = round(data["VALUE"].mean(),2)

mediana_data = data["VALUE"].median()
```

```python
#Saída:
print("Soma:",soma_data, "\nMédia:", media_data, "\nMediana:",mediana_data)
```

```python
#Valor mínimo e máximo de uma coluna:
min_data = data["VALUE"].min()

max_data = data["VALUE"].max()
```

```python
#Saída:
print("Mínimo:",min_data, "\nMáximo:", max_data)
```

```python
#Histograma para o total gasto por município
tot = data["NO_ENTE"].value_counts()

plt.figure(figsize = (10, 7))
x = tot

plt.hist(x, bins = 20, color = "green")
plt.title("Total gasto")
plt.xlabel("Município")
plt.ylabel("Total")
```
