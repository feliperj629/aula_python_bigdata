4) Salário detalhado:

```python
sal_basico = float(input('Digite o salário básico:'))
nrdep = int(input('Digite o número de dependentes:'))
temp = int(input('Digite o tempo de serviço em anos:'))

abono_dependente = nrdep * 50
abono_temp = sal_basico * 0.03 * temp
sal_abono = sal_basico + (abono_dependente + abono_temp)
desconto_inss = sal_abono * 0.05
sal_final = sal_abono - desconto_inss

print('Abono dependente: R$: {:.2f}'.format(abono_dependente))
print('Abono por tempo: R$: {:.2f}'.format(abono_temp))
print('Desconto INSS: R$: {:.2f}'.format(desconto_inss))
print('Salário líquido: R$: {:.2f}'.format(sal_final))
```

5) Fatorial:

```python
'''
Programa que lê um número inteiro n >= 0 e imprime n!
'''
import time
print("Cálculo do fatorial de um número\n")
# leia o valor de n
n = int(input("Digite um número inteiro não-negativo: "))
# inicializações
i = 1  # contador
n_fat = 1  #armazena o valor do fatorial
inicio = time.time()
# calcule n!
while i <= n:
    n_fat = n_fat * i
    i = i + 1
fim = time.time()
total = fim - inicio
print("%d! = %d" %(n, n_fat))
print("Tempo total: " + str(total))
```

12) recebe uma sequência de nrs e verifica se estão ordenados

```python
def checkOrdenacao(lst):
    for i in range(0, len(lst)-1):
        if lst[i] > lst[i+1]:
            return False
    return True
```

14) recebe uma sequência de nrs e verifica se seguem o padrão de incremento

```python
def aritmetica(lst):
    if len(lst) < 2:
        return True
    diff = lst[1] - lst[0]
    for i in range(1, len(lst)-1):
        if lst[i+1] - lst[i] != diff:
            return False
    return True
```

18) cria o acrônimo a partir de um texto informado

```python
def acronimo(frase):
    palavra = frase.split()
    res = ''
    for w in palavra:
        res = res + w[0].upper()
    return res
```

```python
acronimo('random access memory')
```

31) programa para ler o nome e a nota de vários alunos o programa se encerra quando o nome for branco

```python
TURMA = []
while True:
    NOME = input("Nome: ")
    if len(NOME) == 0:
        break
    NOTA = float(input("Nota: "))
    TURMA.append([NOME,NOTA])
    TOTAL = 0
    for ALUNO in TURMA:
        TOTAL = TOTAL + ALUNO[1]
    MEDIA = TOTAL / len(TURMA)
    print("Média:", MEDIA)
    for ALUNO in TURMA:
        if ALUNO[1] > MEDIA:
            print(ALUNO[0], "-", ALUNO[1])
```

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Visualização simples")
plt.show()
```
